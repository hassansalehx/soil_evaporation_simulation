# Soil Evaporation Simulation

**Introduction**  
This notebook demonstrates the simulation of soil evaporation using the soil evaporation capacitance (SEC) model proposed by [Or and Lehman (2019)](https://doi.org/10.1029/2018WR024050). The goal is to simulate how soil moisture evaporates over time following tropical cyclone rainfall under different conditions using a capacitance approach. The model is initialized with moisture content (θ) and based on soil parameters and the potential evaporation, the evaporation rate changes. This notebook imports soil parameters based on the selected soil types that represent the study area taken from [Lehman et al. (2018)](http://dx.doi.org/10.1029/2018GL078803) and shapefiles that contain the initial water content in mm. The code processes the data based on the equations mentioned in detail in the supplementary materials of [Or and Lehman (2019)](https://doi.org/10.1029/2018WR024050), and visualizes the accumulated evaporation volume for each soil type.

This work is part of the study titled **"Intensifying Tropical Cyclone in the Arabian Sea Replenish Depleting Aquifers"** (under review in *Nature Communications Earth and Environment*).

The simulation proceeds by iteratively reducing the soil water content until the soil moisture reaches a residual water content value. It applies different equations during stage 1 where the water content is above the critical water content, and stage 2 when below. The workflow was modified from the appendix of [Or and Lehman (2019)](https://doi.org/10.1029/2018WR024050).

````python
# Import necessary libraries
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from tqdm import tqdm
````

````python
# Define constants
vap_jump = 10 #  ξ, drying front depth
e2 = 1    #  evaporation rates E_2 at the onset of stage-II

# Define function to calculate effective hydraulic conductivity
def calculate_k_eff(ks, alpha, h_crit, n, m):
    """
    Calculate the effective hydraulic conductivity using the given parameters.
    
    Parameters:
      - ks: Saturated hydraulic conductivity.
      - alpha: shape parameter.
      - h_crit: Critical pressure/head.
      - n, pore size distribution
      - m: m=1-1/n.
      
    Returns:
      - k_eff: Effective hydraulic conductivity (converted to mm units).
    """
    k_eff = 4 * ks * np.sqrt((1 + (alpha * h_crit)**n)**(-m)) * (1 - (1 - (1 / (1 + (alpha * h_crit)**n)))**m)**2
    return k_eff * 1000  # Conversion factor to mm
````

````python
# Define the base directory
BASE_DIR = os.getcwd()

# Define the path to the soil parameters Excel file (change the soil types and their parameters as needed)
soil_param_file = os.path.join(BASE_DIR, 'data', 'soil_param.xlsx')

# Load soil parameters from the specified sheet
soil_params = pd.read_excel(soil_param_file, sheet_name='code')
````

## Load Shapefiles of Soil Type distribtuion in the study area

The shapefiles for different soil types (e.g., Dunes, Alluvium, Hard Limestone, Karstified Limestone) are the infiltration output of the hydrodynamic model (RiverFlow2D) for the years 2011, 2018, and 2020 are stored in:

`data/Infiltration_output/<year>/`

The cell below loads these shapefiles into a dictionary for further processing.

````python
# Define the base path for shapefile data
shapefile_base_path = os.path.join(BASE_DIR, 'data', 'Infiltration_output')

# Specify the years and shapefile names to process
years = [2011, 2018, 2020] #event year
shapefile_types = ['Dunes.shp', 'Alluvium.shp', 'Hard Limestone.shp', 'Karstified Limestone.shp']

# Dictionary to store GeoDataFrames for each year and soil type
gdfs = {}

for year in years:
    year_path = os.path.join(shapefile_base_path, str(year))
    gdfs[year] = {}
    for shapefile in shapefile_types:
        shapefile_path = os.path.join(year_path, shapefile)
        # Normalize the key by removing spaces and extension, and converting to lowercase
        key = shapefile.split('.')[0].replace(' ', '_').lower()
        gdfs[year][key] = gpd.read_file(shapefile_path)
        print(f"Loaded shapefile for {year} - {key}")
````

## Simulation of Soil Evaporation

In this section, we define the function `process_gdf` to simulate soil evaporation for each soil type. The simulation works by:

- Iterating through each soil parameter from the Excel file, the parameters are calculated in (Or 2018) supporting information
- For each feature in the GeoDataFrame, computing the evaporation process until the soil moisture falls to the residual level.
- Using two cases:
  - **Stage 1: Water content is above the critical water content threshold:** The evaporation is calculated based on effective hydraulic conductivity.
  - **Stage 2: Water content is below the critical water content threshold:** A different equation is applied.
- Cumulatively summing the evaporation volume.

The resulting evaporation volumes are stored in a dictionary keyed by `<year>_<soil_type>`.

````python
def process_evaporation(gdf):

    # Initialize a list to store DataFrames for each site
    dfs=[]
    sites = []

    # Loop through each row in the dataframe
    for index, row in soil_params.iterrows():

        delta_theta = row['theta_crit'] - row['theta_res'] / 2
        cell_num=0
        dfs=[]
        for s, area in tqdm(zip(gdf['s_mm'], gdf['area_m2']), desc='Processing Sites', total=len(gdf)):
            cell_num+=1
            theta = s/(row['lc'] * 1000)
            if theta >= row['theta_sat']:
                theta = row['theta_sat']
                s=theta*(row['lc'] * 1000)
            elif theta <=row['theta_res']:
                s=0
                continue
            else:
                s = s
            # Soil evaporation process
            day = 0
            days2 = 0
            t2 = 0
            cumulative_es = 0  # Initialize cumulative es
            cumulative_vol=0
            cum_s = s
            es=0
            q=0
            q_current=0
            Es_list=[]
            es_current=0
            while theta > row['theta_res']:
                if theta > row['theta_crit']:
                    day += 1
                    t2 = day
                    theta-=0.01
                    q_pre=q
                    theta_eff = (theta - row['theta_res']) / (row['theta_sat'] -
                                                              row['theta_res'])
                    k_eff = calculate_k_eff(row['ks'], row['alpha'], row['h_crit'], row['n'], row['m'])

                    k_theta = row['ks'] * 1000 * (theta_eff)**(0.5) * ( 1 - (1 - theta_eff**(1 / (1 - (1 / row['n']))))**(1 - (1 / row['n'])))**2
                    es_current = (row['et'] * k_theta * (1 + (row['et'] / k_eff))) / (row['et'] + (k_theta * (1 + (row['et'] / k_eff))))
                    print()
                    if (cum_s-es_current)/ (row['lc'] * 1000) > row['theta_crit']:
                        q=k_theta *np.exp(-(day*k_theta)/((row['lc'] * 1000)*(theta-row['theta_crit'])))
                        q_current=q-q_pre

                else:
                    q_current=0
                    es_previous=es
                    days2 += 1
                    t = t2 + days2
                    es = np.sqrt(vap_jump) / np.sqrt(vap_jump + (2 * e2 * (t - t2)) / delta_theta) * (  2 * e2 * (t - t2) + (delta_theta * vap_jump)) - ( vap_jump * delta_theta)
                    es_current=es - es_previous

                cum_s -= (es_current+q_current)
                es_vol=es_current*(10**-6)*area*(10**-6)
                cumulative_vol +=es_vol
                theta = cum_s / (row['lc'] * 1000)
                Es_list.append(cumulative_vol)

            # Create a DataFrame for the current cell
            df = pd.DataFrame({cell_num: Es_list})

            # Append the DataFrame to the list
            dfs.append(df)


        soil_evaporation_df = pd.concat(dfs, axis=1)
        soil_evaporation_df_filled = soil_evaporation_df.ffill( axis=0)
        sum_columns = soil_evaporation_df_filled.sum(axis=1)
        df_site = pd.DataFrame({row['site']: sum_columns})
        sites.append(df_site)
    soil_evaporation_vol = pd.concat(sites, axis=1)
    return soil_evaporation_vol
````

````python
# Dictionary to store the processed DataFrames
# Note: if you find that the processing bar makes funny stuff it's fine as long as no errors show up and you can generate the plots

processed_soil_evap_vol = {}

# Iterate through each year and each soil type
for year in years:
    for soil_type, gdf in gdfs[year].items():
        # Create the variable name dynamically using both year and soil type
        key = f"{year}_{soil_type}"  # Example: "2011_dunes"
        processed_soil_evap_vol[key] = process_evaporation(gdf)
````

````python
# Define soil type mappings based on location names
soil_type_param_map = {
    "dunes": "USMo1",
    "karstified_limestone": "USMo1",
    "alluvium": "USMo7",
    "hard_limestone": "USDk1"
}

# Create a new dictionary to store only the mapped values
subset_soil_evap_vol = {}

# Iterate over the processed dataset
for key, df in processed_soil_evap_vol.items():
    # Extract year and soil type
    year, soil_type = key.split('_', 1)  # Example: "2011_dunes" → year="2011", soil_type="dunes"
    
    # Ensure lowercase for mapping consistency
    soil_type = soil_type.lower()

    # Check if the soil type exists in our mapping
    if soil_type in soil_type_param_map:
        param_column = soil_type_param_map[soil_type]  # Get the corresponding parameter

        # Check if the column exists in the dataframe before subsetting
        if param_column in df.columns:
            subset_soil_evap_vol[key] = df[[param_column]]  # Keep only the mapped column
            print(f"Subsetting {key}: Keeping column {param_column}")
        else:
            print(f"WARNING: Column {param_column} not found in {key}, skipping.")
    else:
        print(f"WARNING: No parameter mapping found for {soil_type}, skipping.")

# The new dictionary `subset_soil_evap_vol` now contains only the relevant columns for each soil type.
````

````python
def prepare_evaporation_data(subset_soil_evap_vol):
    """
    Extracts relevant soil evaporation data and computes total evaporation fro each event.
    This version handles columns with different lengths by forward-filling each column
    to match the length of the longest column before summing.

    Parameters:
        - subset_soil_evap_vol: Dictionary containing DataFrames with processed evaporation values for each year and soil type.

    Returns:
        - processed_data: Dictionary where keys are years and values are DataFrames with:
            - Individual soil type evaporation (forward-filled to the longest length)
            - A 'Total' column summing across all soil types
    """
    processed_data = {}

    # Get unique years from the dataset
    years = sorted(set(int(key.split('_')[0]) for key in subset_soil_evap_vol.keys()))

    for year in years:
        # Filter data for the current year
        year_keys = [key for key in subset_soil_evap_vol.keys() if key.startswith(str(year))]
        if not year_keys:
            print(f"WARNING: No data found for year {year}. Skipping processing.")
            continue
        
        # Create a dictionary with soil types as keys and their respective DataFrames
        year_data = {key.split('_')[1]: subset_soil_evap_vol[key] for key in year_keys}

        # Determine the maximum length among the soil type DataFrames
        max_length = max(df.shape[0] for df in year_data.values())

        # Create an empty DataFrame with index range equal to the maximum length
        total_evap_df = pd.DataFrame(index=range(max_length))

        # Loop through each soil type and reindex (forward fill) its data to the max length
        for soil_type, df in year_data.items():
            if df.empty:
                continue
            # Reindex the DataFrame to the maximum length and forward fill missing values
            df_reindexed = df.reindex(range(max_length)).ffill()
            # Add the reindexed column to the total evaporation DataFrame
            total_evap_df[soil_type] = df_reindexed.iloc[:, 0]
        
        # Compute the 'Total' column as the row-wise sum of all soil types
        total_evap_df['Total'] = total_evap_df.sum(axis=1)
        processed_data[year] = total_evap_df

    return processed_data
#run the function
processed_data = prepare_evaporation_data(subset_soil_evap_vol)
````

## Visualization Using Matplotlib

````python
def plot_soil_evaporation(processed_data):
    """
    Plots soil evaporation for each year as three subplots, one for each year.
    Each subplot includes individual soil type evaporation and a total evaporation line.

    Parameters:
        - processed_data: Dictionary with yearly DataFrames containing:
            - Soil type evaporation
            - Total evaporation sum
    """
    
    # Define years and labels
    years = sorted(processed_data.keys())
    labels = ['(a)', '(b)', '(c)']

    # Create subplots
    fig, axes = plt.subplots(1, len(years), figsize=(16, 5), gridspec_kw={'wspace': 0.2})  # One row, dynamic columns
    
    for idx, year in enumerate(years):
        ax = axes[idx]  # Select subplot for the current year
        df = processed_data[year]  # Get DataFrame for this year

        if df is None or df.empty:
            print(f"WARNING: No data found for year {year}. Skipping plot.")
            continue
        
        # Plot individual soil types
        for col in df.columns[:-1]:  # Exclude 'Total' column
            ax.plot(df.index, df[col], label=col.title())

        # Plot total evaporation as a separate line
        ax.plot(df.index, df['Total'], label='Total', linestyle='--', linewidth=2, color='black')

        # Set labels and aesthetics
        ax.set_xlabel('Days', fontsize=15)
        ax.set_xlim(0, len(df))
        max_y_value = df.iloc[:, 1:6].max().max()  # Find the maximum value across all plotted columns
        ax.set_ylim(0, max_y_value+1)  # Set y-axis from 0 to the max value  

        if idx == 0:
            ax.set_ylabel('Accumulated Soil Evaporation (km$^3$)', fontsize=15)
            ax.legend(loc='best', fontsize=12)

        ax.tick_params(axis='x', rotation=45, labelsize=14)
        ax.tick_params(axis='y', labelsize=14)
        
        # Add subplot label (a), (b), (c)
        ax.text(0.98, 0.9, labels[idx], transform=ax.transAxes, fontsize=16, fontweight='bold',
                va='center', ha='right', bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
        
        # Calculate difference between last value and value at day 100
        # this is for veryfying how much of the water content will evaporate within a few months and how much would remain
        if len(df) > 100:
            last_value = df['Total'].iloc[-1]
            value_at_day_100 = df['Total'].iloc[99]
            difference = last_value - value_at_day_100
            print(f"Difference between last value and day 100 for {year}: {difference:.3f} km³")

    plt.show()

#run the function
plot_soil_evaporation(processed_data)
````
